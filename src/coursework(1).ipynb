Outlier engineering

```python
def find_skewed_boundaries(df, variable, distance):
    IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)

    lower_boundary = df[variable].quantile(0.25) - (IQR * distance)
    upper_boundary = df[variable].quantile(0.75) + (IQR * distance)

    return upper_boundary, lower_boundary
```

```python
Schooling_upper_limit, Schooling_lower_limit = find_skewed_boundaries(df, 'Schooling', 2)
Schooling_upper_limit, Schooling_lower_limit
```

```python
Income_composition_of_resources_upper_limit, Income_composition_of_resources_lower_limit = find_skewed_boundaries(df, 'Income_composition_of_resources', 1.5)
Income_composition_of_resources_upper_limit, Income_composition_of_resources_lower_limit
```

```python
df.columns
```

```python
X = df[[ 'Adult_Mortality', 'Alcohol', 'percentage_expenditure', 'Hepatitis_B',
       'Measles', 'BMI', 'under-five_deaths', 'Polio', 'Total_expenditure',
       'Diphtheria', 'HIV/AIDS','Population', 'thinness__1-19_years',
       'thinness_5-9_years', 'Income_composition_of_resources', 'Schooling']]

y = df['Life_expectancy']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2004)
```

```python
from sklearn.svm import SVR
import time
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error,r2_score, mean_absolute_error
models = {
'Linear Regression': LinearRegression(),
 'SVR': SVR(),
  'XGBoost': XGBRegressor(),
  'KNeighbors': KNeighborsRegressor(),
'Decision Tree Regressor': DecisionTreeRegressor(),
'Random Forest Regressor': RandomForestRegressor(),
'Ridge Regression': Ridge()
}

results = []

for model_name, model in models.items():
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    results.append({
    'Model name': model_name,
    'Mean Absolute Error': mae,
    'Mean Squared Error': mse,
    'R2 Score': r2})
```

```python
from sklearn.model_selection import ShuffleSplit

ss = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
for train_index, test_index in ss.split(X_train):
    X_train_fold, X_test_fold = X_train.iloc[train_index], X_train.iloc[test_index]
    y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]

    model = RandomForestRegressor()
    model.fit(X_train_fold, y_train_fold)
    y_pred = model.predict(X_test_fold)

    mse = mean_squared_error(y_test_fold, y_pred)
    mae = mean_absolute_error(y_test_fold, y_pred)
    r2 = r2_score(y_test_fold, y_pred)
    print("----")
    print('MSE: ', mse)
    print('MSE: ', mae)
    print('MSE: ', r2)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': np.arange(25,55,10),
    'max_depth': [10, 15, 20],
    'min_samples_split': [5,10,15,20],
    'min_samples_leaf': [2,3,4],
    'bootstrap':[True,False]
}

rf = RandomForestRegressor()

grid = GridSearchCV(estimator=rf,
                    param_grid=param_grid,
                    scoring='r2',
                    cv=5,
                    verbose=3,
                    return_train_score=True)

grid.fit(X_train, y_train)

```

```python
print("The best parameters are %s with a R2 of %0.2f"
      % (grid.best_params_, grid.best_score_))
```
